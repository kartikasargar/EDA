```python
df.dropna(inplace=True)
```

```python
df.fillna(method="ffill",inplace=True)
```

```python
df.dtypes
```

```python
print(df.shape)
df.head()
```

```python
df["engine_Group"]=pd.cut(
df["engine_capacity(CC)"],
bins=[0,1200,1800,5000],
labels=["Small","Medium","Large"])
```

```python
brand_encoded=pd.get_dummies(df["brand"])
brand_encoded
```

```python
df["Combined"]=df["price"]/df["overall_cost"]
df["Combined"]
```

```python
def price_category(price):
 if price<=800000:
  return"Low"
 elif price<=1000000:
  return "Medium"
 else:
  return "High"
df["price_category"]=df["price"].apply(price_category)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
df["make_year_scaled"]=Scaler.fit_transform(df[["make_year"]])
df["price_scaled"]=Scaler.fit_transform(df[["price"]])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
df["make_year_std"]=Scaler.fit_transform(df[["make_year"]])
df["price_std"]=Scaler.fit_transform(df[["price"]])
df
```
